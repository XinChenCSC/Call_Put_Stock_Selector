```python
import pandas as pd
import datetime as dt
import yfinance as yf
from pandas_datareader import data as pdr

yf.pdr_override()
start = dt.datetime(2022, 7, 1)
now = dt.datetime.now()

filepath="sp500.csv"
stocklist = pd.read_csv(filepath)
exportList = []
putList = []

for i in stocklist.index:
    stock=str(stocklist["Symbol"][i])
    
    try:
        df = pdr.get_data_yahoo(stock,start,now)
        low_list = df['Low'].rolling(14, min_periods=14).min()
        low_list.fillna(value=df['Low'].expanding().min(), inplace=True)
        high_list = df['High'].rolling(14, min_periods=14).max()
        high_list.fillna(value=df['High'].expanding().max(), inplace=True)
        
        rsv = (df['Close']-low_list) / (high_list - low_list) * 100.0
        df['K'] = rsv.ewm(com=2, adjust=False).mean()
        df['D'] = df['K'].ewm(com=2, adjust=False).mean()

        diff = df['Close'].ewm(span=12, adjust=False).mean()-df['Close'].ewm(span=26, adjust=False).mean()
        dea = diff.ewm(span=9, adjust=False).mean()
        df['MACD'] = 2*(diff-dea)
        
        if (df['MACD'][-1] <0 and df['K'][-1]>df['D'][-1] and not(df['MACD'][-2] <0 and df['K'][-2]>df['D'][-2]) ):
            exportList.append(stock)
        if (df['MACD'][-1] >0 and df['K'][-1]<df['D'][-1] and not(df['MACD'][-2] <0 and df['K'][-2]<df['D'][-2]) ):
          putList.append(stock)
    except Exception:
        print("No data on " + stock)

if len(exportList) == 0:
    print("No stock can buy")
else:
    print("{call}")
    index = 0
    for j in exportList:
        print(str(index) + ". "+ j)
        index+=1
      
if len(putList) == 0:
    print("No stock can put")
else:
    print("{put}")
    index = 0
    for j in putList:
        print(str(index) + ". "+ j)
        index+=1
```
